# Cardiotocography Data Set - Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Configure Matplolib

```python
# To use with Pycharm and dark theme
plt.style.use('ggplot')

#to increase font size
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 27}

plt.rc('font', **font)
```

## Read dataset, clean and preprocess

```python
ctg = pd.read_excel('./dataset/CTG.xls', 'Raw Data')
```

```python
X_ctg = ctg.iloc[1:2127, 6:-12]
X_ctg.drop('DR', axis=1, inplace=True)

y3_ctg = ctg.iloc[1:2127, -1]
y10_ctg = ctg.iloc[1:2127, -2]
```

```python
X_ctg.describe()
```

## Plot histograms


```python
fig = plt.figure(dpi=150, figsize=(30, 30))
ax = fig.gca()
X_ctg.hist(ax = ax)
plt.savefig('histogram.png')
plt.show()
```

```python
y3_ctg.hist()
plt.savefig('y3_histogram.png')
plt.show()
```

```python
y10_ctg.hist()
plt.savefig('y10_histogram.png')
plt.show()
```

## Plot boxplots


```python
#select list of features with similar ranges
big_features = ['LB', 'ASTV', 'Width', 'Min', 'Max', 'Mode', 'Mean', 'Median', 'ALTV', 'FM', 'Variance']
small_features = X_ctg.columns.difference(big_features)
```

```python
fig = plt.figure(dpi=100, figsize=(30, 30))
ax = fig.gca()
X_ctg[big_features].boxplot(ax = ax)
plt.savefig('boxPlot1.png')
plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(30, 30))
ax = fig.gca()
X_ctg[small_features].boxplot(ax = ax)
plt.savefig('boxPlot2.png')
plt.show()
```
