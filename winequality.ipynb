# WineQuality Data Set - Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Configure Matplolib

```python
# To use with Pycharm and dark theme
plt.style.use('ggplot')

#to increase font size
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 27}

plt.rc('font', **font)
```

## Read dataset, clean and preprocess

```python
ctg_white = pd.read_csv('./dataset/winequality-white.csv', sep =";")
#ctg_red = pd.read_csv('./dataset/winequality-red.csv', sep =";")
#ctg = pd.concat([ctg_white, ctg_red], axis = 0)
ctg = ctg_white
```

```python
ctg.iloc[:].describe()
```

```python
X_ctg = ctg.iloc[:, 0:-1]

y7_ctg = ctg.iloc[:, -1].to_frame()
```

```python
X_ctg.describe()
```

```python
y7_ctg.describe()
```

## Plot histograms


```python
fig = plt.figure(dpi=150, figsize=(30, 30))
ax = fig.gca()
X_ctg.hist(ax = ax)
plt.savefig('histogram.png')
plt.show()
```

```python
y7_ctg.hist()
plt.savefig('y3_histogram.png')
plt.show()
```

```python
y7_ctg.boxplot()
plt.savefig('y3_histogram.png')
plt.show()
```

## Plot boxplots


```python
#select list of features with similar ranges
big_features = ['free sulfur dioxide', 'total sulfur dioxide']
small_features = X_ctg.columns.difference(big_features)
```

```python
fig = plt.figure(dpi=100, figsize=(30, 30))
ax = fig.gca()
X_ctg[big_features].boxplot(ax = ax)
plt.savefig('boxPlot1.png')
plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(30, 30))
ax = fig.gca()
X_ctg[small_features].boxplot(ax = ax)
plt.savefig('boxPlot2.png')
plt.show()
```
